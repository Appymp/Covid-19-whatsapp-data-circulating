```python
df=pd.read_csv("BlrQuar.csv")
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
#Cleanup column names
df.columns=df.columns.str.replace(' ','_')
df.columns=df.columns.str.replace('\r','_')
df = df.rename(columns={'Date_until_Quarantine_d_at_home': 'Date_until_Quarantined_at_home',
                        'Street/_Village': 'Street_Village', 
                        'District/_City' : 'District_City',  
                        'Final_District/_City' : 'Final_District_City'
                       })
df.columns
```

```python
#Check which columns have how many missing values
df.isnull().sum()
```

```python
# View each of the columns with missing values
df[df.House_No.isnull()]
```

```python
# Handling the missing data:
df.dropna(subset=['Port_of_Origin_of_journey'],inplace=True) #remove missing values in this column
df['Port_of_Final_Destination'].fillna('Bangalore',inplace=True) #must be bangalore right?
```

```python
# Salvage PIN from 'State' column
df_pin=df[pd.to_numeric(df['State'], errors='coerce').notnull()] # 455 rows initial
for ind,row in df_pin.iterrows():
    df.loc[ind,'Tehsil':]=df.loc[ind,'Tehsil':].shift(1) #shift row contents by +1 column
df[pd.to_numeric(df['State'], errors='coerce').notnull()] #check
```

```python
# Check for PIN in Final district city column
df_fdc=df[pd.to_numeric(df['Final_District_City'], errors='coerce').notnull()] #378 rows initially
for ind,row in df_fdc.iterrows():
    df.loc[ind,:'Final_District_City']=df.loc[ind,:'Final_District_City'].shift(-1) #shift row contents by -1 column
df[pd.to_numeric(df['Final_District_City'], errors='coerce').notnull()] #check
```

```python
df[df.PIN.isnull()] #123 rows of null pin values. Can be dropped.
df.dropna(subset=['PIN'],inplace=True)
df[pd.to_numeric(df['PIN'], errors='coerce').isnull()] # 1409 non-numeric values. No PIN info salvageable. So remove
df=df[~pd.to_numeric(df['PIN'], errors='coerce').isnull()] # non numeric removed
df[pd.to_numeric(df['PIN'], errors='coerce').isnull()] #check
```
